##5
모든 정점 사이에 간선이 존재하는 그래프라면 인접 행렬을 사용할 것이다. 행렬은 간선의 존재 여부를 매우 빠른 시간 복잡도로 확인 가능하기 때문이다. 모든 정점에 간선이 존재할 때 인접 리스트를 사용한다면 매우 많은 시간이 소요될 것이다.

##6
인접 리스트를 사용할 것이다. 간선이 드물게 존재하는 그래프에서 인접 행렬을 이용한다면 불필요한 정보가 많이 들어가 메모리의 낭비가 심할 것이다. 인접 리스트릴 이용한다면 실제 존재하는 간선 정보만 저장하므로 메모리 공간을 행렬에 비해 절약할 수 있다.

##16
Topological Sort with STACK:
 Create STACK
 Create visited <- false
 for each vertex v
  if (!visited [v])
   DFS (v, visited, STACK)
 while (!STACK.empty)
   STACK.pop
  DFS (v, visited, STACK )
  visited[v] <- true
  for i in Adj [v]
   if (!visited [i])
    DFS STACK.push(v)
